```python
import pickle
import joblib

# pre-processed text
corpus = {} 

# file to document number mapping
document_map = {}

# key and corresponding posting list
dictionary = {}

with open('/content/DocTerms_SpacyLemm.pickle', 'rb') as handle:
  corpus = pickle.load(handle)
```

```python
corpus['pussboot.txt']
```

```python
len(corpus.keys())
```

```python
counter = 0
for key in corpus.keys():
  if (key not in document_map.keys()):
    counter += 1
    document_map[key] = counter
```

```python
for key in corpus.keys():
  list_of_words = corpus[key][0]
  for word in list_of_words:
    if (word[0] in dictionary.keys()):
      dictionary[word[0]][0] = dictionary[word[0]][0] + 1
      if document_map.get(key) in dictionary[word[0]][1]:
        dictionary[word[0]][1][document_map.get(key)].append(word[1])
      else:
        dictionary[word[0]][1][document_map.get(key)] = [word[1]]
    else :
      # dictionary[word[0]] = [[],[document_map.get(key)]]
      dictionary[word[0]] = []
      dictionary[word[0]].append(1)
      dictionary[word[0]].append({})     
      dictionary[word[0]][1][document_map.get(key)] = [word[1]] 
```
